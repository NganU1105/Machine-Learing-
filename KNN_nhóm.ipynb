```python
import seaborn as sns
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import os 
os.chdir("E:\mining\DATA SET")
```

```python
data = pd.read_excel("BikeBuyer.xlsx")
```

```python
data.sample(10)
```

```python
data.Gender = data.Gender.map({'Male':0, 'Female':1}) 
```

```python
data.HomeOwner = data.HomeOwner.map({'No':0, 'Yes':1}) 
```

```python
data.BikeBuyer = data.BikeBuyer.map({'No':0, 'Yes':1}) 
```

```python
data.sample(10)
```

```python
x = data[['YearlyIncome','HomeOwner']]
y = data['BikeBuyer']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
print('Training size:', len(y_train))
print('Test size:', len(y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = )
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
```

```python
from sklearn import metrics
print("Độ chính xác của mô hình :",metrics.accuracy_score(y_test, prediction))
```

```python
y_pred=knn.predict(x_test)
```

```python
KNN = pd.DataFrame({'True values ':y_test ,'y_pred':y_pred})
KNN.sample(5)
```

```python
clf3 = KNeighborsClassifier(n_neighbors= 3, p=2)
clf3.fit(x_train, y_train)
y_pred3 = clf3.predict(x_test)
print('Result for first 20:')
print('Predicted labels:', y_pred3[0:30].tolist())
print('Ground truth:    ', y_test[0:30].tolist())
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
print(" {} nn score: {} ".format(19,knn.score(x_test,y_test)))
```

```python
score_list = []
for each in range(1,25):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
plt.figure(figsize=(8,5))    
plt.plot(range(1,25),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```
